```python
webPageUrl =requests.get('https://en.wikipedia.org/wiki/List_of_terrorist_incidents')
allUrlList = soup(webPageUrl.text,'html.parser')
allUrl = allUrlList.find_all('div',attrs={'class':'div-col columns column-width'})
listOfTerroristIncidentsWrtTime=pd.DataFrame()
for each in allUrl:
    j= each.find_all('a',attrs={'href':True})
    for elem in j:
        listOfTerroristIncidentsWrtTime = pd.concat([listOfTerroristIncidentsWrtTime,
                                                     pd.Series('https://en.wikipedia.org/'+elem['href'])],axis=0)
    break
listOfTerroristIncidentsWrtTime.columns=['Year_of_Terrorist_attack']
listOfTerroristIncidentsWrtTime.reset_index(drop=True,inplace=True)
```

```python
def func(x):
    newX=''
    if str(x).find('(') >0:
        #print('ssss: ',x)
        newX= str(x).replace('(s)','s')
        #print('newX: ',newX)
    elif str(x) == 'Perpetrator':
        newX = 'Perpetrators'
    elif str(x) == 'Deaths':
        newX = 'Dead'
    elif str(x) == 'Injuries':
        newX = 'Injured'
    elif str(x) == 'Dates':
        newX = 'Date'
    else:
        newX = x
    return newX
```

```python
def tagNames(tags,each):
    j=pd.DataFrame()
    k=pd.DataFrame()
    h= requests.get(each)
    weburl= soup(h.text,'html.parser')
    headersList = pd.Series()
    for each in tags:
        thList=each.find_all('th')
    for elem in thList:
        headersList = headersList.append(pd.Series(elem.text.strip()))
    headersList = headersList.unique()
    headersList = pd.Series(headersList).apply(lambda x : func(x))
    headersList = headersList.unique()
    for each in tags:
        for elem in each.find_all('tr'):
            for each1 in elem.find_all('td'):
                j = pd.concat([j,pd.Series(each1.text.strip())],axis=1)
                if j.shape[1] == headersList.size:
                     break
            while j.shape[1] <  headersList.size:
                j = pd.concat([j,pd.Series(' ')],axis=1)
            k =pd.concat([k,j],axis=0)
            j=pd.DataFrame()
    k.columns=[headersList]
    k = k.iloc[1:,:]
    return(k)
    
```

```python
mergedDataFrame=pd.DataFrame()
for each in listOfTerroristIncidentsWrtTime.Year_of_Terrorist_attack:
        #print(each)
        h= requests.get(each)
        weburl= soup(h.text,'html.parser')
        tags = weburl.find_all('table',attrs={'class':'wikitable sortable'})
        tags1 = weburl.find_all('table',attrs={'class':'wikitable'})
        if len(tags) == 0:
            tags = tags1
        allTagNames= tagNames(tags,each)
        mergedDataFrame = pd.concat([mergedDataFrame,allTagNames],axis=0,ignore_index=True)
#display(mergedDataFrame)
        
    
```
